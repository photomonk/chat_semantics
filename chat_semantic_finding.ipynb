```python
def generate_document(chat_data):
    user_documents={user:[] for user in ["Shlok","Ram","Bob","John"]}
    for chat in  chat_data:
        if chat['person']in user_documents:
            user_documents[chat['person']].append(chat)
    return user_documents        
```

```python
def generate_chat_data(num_messages):
    rd.seed('1234')
    chat_data=[]
    users=["Shlok","Ram","Bob","John"]
    start_time=datetime(2025,12,1)

    for  i in range(num_messages):
        timestamp=start_time + timedelta(minutes=5 *i) 

        person=rd.choice(users)
       
        topic =rd.choice(list(sample_messages.keys())) 
        # selection of the topic based on the list of keys in the dic created sample_message 
        
        message=rd.choice(sample_messages[topic])
        # for that selected topic then the message is selection randomly 


        chat_data.append({
            "timestamp":timestamp,
            "person":person,
            "topic":topic,
            "message":message,
        })  ## the list contains the dictonary as the element 
        
    return chat_data
```

```python
chat_data=generate_chat_data(1000)
```

```python
client = PersistentClient(path="./chat_DB")
```

```python
chat_data=client.delete_collection("chat_data")
```

```python
chat_data=client.create_collection("chat_data"
                                   ,metadata={        "hnsw:space": "cosine",           # cosine similarity
        "hnsw:construction_ef": 200,      # higher = better index quality, slower build
        "hnsw:search_ef": 90,             # higher = more accurate queries
        "hnsw:M": 32                      # number of edges per node
                                             }
                                                      )
```

```python
indexes=[]
for index,date in enumerate(data['timestamp']):
    indexes.append(index)
```

```python
chat_data.add(
    documents=[message for message in data['message']],
      embeddings=embeddings['dense_vecs'],
    ids=[str(index) for index in indexes],
    metadatas=[{
        "timestamp":time,
        "person":pe,
    }for time,pe in zip(data['timestamp'],data['person'])]
    
)
```
